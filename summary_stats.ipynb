```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('subset_data-multi-value-traits.csv', index_col=False)  
```

```python
genotype = []
ids = dataframe['plantbarcode'].tolist()

for i in ids:
    if str(i)[3:5] == "22":
        genotype.append("Atlas")
    elif str(i)[3:5] == "16":
        genotype.append("PI_195754")
dataframe["genotype"] = genotype
```

```python
angle_df = dataframe.loc[dataframe['trait'] == 'segment_insertion_angle']
relevant_data = [pd.to_numeric(angle_df["value"]), angle_df["genotype"]]
headers = ["value", "genotype"]
df = pd.concat(relevant_data, axis=1, keys=headers)

grouped = df.groupby('genotype')['value'].mean().plot(kind='barh', title='Mean Leaf Angle in Degrees')
plt.show()
```
